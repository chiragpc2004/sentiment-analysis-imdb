## **Loading the IMDB 50K Movie Reviews Dataset and Performing Basic Exploratory Data Analysis (EDA)**

### **Exploratory Data Analysis**

- Understand the **structure of the dataset** including rows, columns, and data types.
- Gain **basic insights** using `describe()` and `info()` to summarize the data.
- Analyze the **count of unique values**, check for **duplicates**, and identify any **missing values**.
- Explore the **number of words and characters** in each review to understand the text length distribution.
- Perform **basic exploratory analysis** to uncover patterns or anomalies in the dataset.


#### Import statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
df = pd.read_csv("D:/imdb-sentiment-classifier/data/raw/IMDB_Dataset.csv")
df.head()
```

```python
# Some basic information
display(df.info())
```

```python
# Some insights on the data
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Number of distinct values in sentiment column
df['sentiment'].value_counts()
```

```python
# Number of distinct reviews
print("Total number of distinct reviews: ",df['review'].value_counts().count())
```

```python
# Check for duplicates
print("Total number of duplicated reviews:", df.duplicated().sum())
```

```python
# Sentiment bar plot
sns.countplot(x='sentiment', data=df)
plt.title("Distribution of Sentiment Labels")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
# Distribution of review lengths
review_lengths = df['review'].apply(lambda x: len(x.split()))
plt.hist(review_lengths, bins=50, color='salmon')
plt.title('Distribution of Review Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# Boxplot of Review Length by sentiment
df['review_length'] = df['review'].apply(lambda x: len(x.split()))
sns.boxplot(x='sentiment', y='review_length', data=df)
plt.title('Review Length by Sentiment')
plt.xlabel('Sentiment (0 = Negative, 1 = Positive)')
plt.ylabel('Review Length (Words)')
plt.show()
```

```python
# Review length stats
df['char_length'] = df['review'].str.len()
df['word_count'] = df['review'].str.split().str.len()
print("\nCharacter length stats:\n", df['char_length'].describe())
print("\nWord count stats:\n", df['word_count'].describe())
```
